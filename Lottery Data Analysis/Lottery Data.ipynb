## Finding most frequent numbers in 5 sets and their probabilites from April 7 to July 7(3 months period)

### Set 1 count and probabilities

```python
count_1 =df[0].value_counts().reset_index()

count_1.columns = ['num_set', 'Occurences']

count_1['Probabilities occur(%) in set'] = round(count_1['Occurences']/len(count_1)*100,1)

count_1
```

```python
figure(figsize =(12,8))
plt.title('Number occurences in set 1',fontsize=20)
sns.barplot(x= 'num_set', y = 'Occurences', data = count_1)
plt.xlabel('Number', fontsize=18)
plt.ylabel('Occurences', fontsize=16)
```

### Probabilities as a whole

```python
set_1 = df1[0].value_counts().reset_index()
set_1.columns = ['num_set', 'Occurences']

set_2 = df1[1].value_counts().reset_index()
set_2.columns = ['num_set', 'Occurences']

set_3 = df1[2].value_counts().reset_index()
set_3.columns = ['num_set', 'Occurences']

set_4 = df1[3].value_counts().reset_index()
set_4.columns = ['num_set', 'Occurences']

set_5 = df1[4].value_counts().reset_index()
set_5.columns = ['num_set', 'Occurences']

df2 = pd.concat([set_1,set_2,set_3,set_4,set_5])

```

```python
df3 =df2.groupby('num_set').sum()['Occurences'].reset_index()

df3['Probabilities(%)'] = round(df3['Occurences']/len(df3),3)*100
```

```python
rank =df3.sort_values(by = 'Probabilities(%)',ascending = False).head(20)

rank
```

```python
figure(figsize =(15,8))
plt.title('Number occurences of winning tickets in 3 months',fontsize=20)
sns.barplot(x= 'num_set', y= 'Occurences', data = rank)
plt.xlabel('Number', fontsize=18)
plt.ylabel('Occurences', fontsize=16)
```

## Summary finding lottery (April 7,2022 to July 7 , 2022)

* You have higher chance to win at least 2 matching numbers if you have  number (39,38,35,34,27) as your set 5 and number(2,5,4,3,1) as your set 1 in your ticket(around 51% average chance of winning a free play).

* Having number (39) as set 5 and number(2 and 5) as set 1 will almost guarantee a a chance to win free play and increase your odds winning higher prizes(with probability of 70% winning free play and could be higher than 40% winning at least 3 matching prizes)

* Set (2, 3 and 4) are more randomize than (1 and 5) with most probability occur in set less than 50%.

* You are more likely to win if you have the top 5 number as a whole (8,39,27,22, and 10) in any set which will increasae your chance of winning by approximately 41%.
